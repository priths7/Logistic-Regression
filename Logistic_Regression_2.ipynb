```python
embarked = pd.get_dummies(data['Embarked'],drop_first=True)
embarked.head(5)
```

```python
pclass = pd.get_dummies(data['Pclass' ],drop_first=True)
pclass.head(5)
```

```python
new_data = pd.concat([data,gender,embarked,pclass],axis=1)
new_data.head(5)
```

```python
new_data.head(5)
```

```python
new_data.drop(['Pclass','Name','Sex','Embarked','Ticket'],axis=1,inplace=True)
new_data.head(5)
```

# Train and Test

---



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score,log_loss

x = new_data.drop('Survived',axis=1)
y = new_data['Survived']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=1)
```

**Data Data Preprocessing using sklearn preprocessing library**

Transform features by scaling each feature to a given range


```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
x_train = mm.fit_transform(x_train)
x_test = mm.fit_transform(x_test)
```

**Model Training**

```python
logmodel = LogisticRegression(solver='lbfgs')
logmodel.fit(x_train,y_train)
predict = logmodel.predict(x_test)
print('Classification Report')
print(classification_report(y_test,predict))
```

# Confusion Matrix
Confusion Matrix is a table like representation or a matrix which consists of True Negative , False Positive , True Positive, False Negative. This is used to find Precision , Recall and F1 Score of the model

```python
print('Confusion Matrix')
print(confusion_matrix(y_test,predict))
```

```python
print('Accuracy Report')
print(accuracy_score(y_test,predict))
```

```python
print('f1_score')
print(f1_score(y_test,predict))
```

```python
print('log loss score')
print(log_loss(y_test,predict))
```
